```python
sns.barplot(x="gender",y = "math score",data=df)
plt.title('gender vrs avg_maths score')
```

```python
sns.scatterplot(x="math score", y="reading score", data=df)
plt.title('Scatter Plot with Color by Time')
```

```python
sns.scatterplot(x="writing score", y="reading score", data=df)
plt.title('Scatter Plot with Color by Time')
```

```python
sns.barplot(x="test preparation course",y = "math score",data=df)
plt.title('test preparation vrs math score')
```

```python
sns.histplot(df["math score"],kde=True)
plt.title("math score distribution")
```

```python
sns.countplot(x="gender",data=df)
plt.title('counts of students by gender')
```

```python
sns.violinplot(x='gender', y='writing score', data=df)
plt.title('Gender vs Writing Score')
plt.show()
```

```python
sns.scatterplot(x='math score', y='reading score', data=df)
plt.title('Math vs Reading Score')
```

```python
sns.barplot(
    x='gender',
    y='math score',
    hue='test preparation course',
    data=df
)
plt.title('Gender & Test Prep vs Math Score')
plt.show()
```
