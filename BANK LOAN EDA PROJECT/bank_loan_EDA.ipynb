# Importing libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

### loading datasets

```python
customers = pd.read_csv("customers.csv")
loans = pd.read_csv("loans.csv")
payments = pd.read_csv("payments.csv")
branches = pd.read_csv("branches.csv")
```

### Creating a dataset dictionary

```python
datasets = {
    "Customers": customers,
    "Loans": loans,
    "Payments": payments,
    "Branches": branches
}
```

### Missing Values

```python
for name, df in datasets.items():
    print(f"\n{name}")
    print(df.isnull().sum())
```

```python
customers["annual_income"] = customers["annual_income"].fillna(customers["annual_income"].median())
```

```python
customers["education"] = customers["education"].fillna(customers["education"].mode()[0])
```

```python
customers["marital_status"] = customers["marital_status"].fillna(customers["marital_status"].mode()[0])
```

```python
payments["amount_paid"] = payments["amount_paid"].fillna(0)
```

```python
customers["cibil_score"] = customers["cibil_score"].fillna(customers["cibil_score"].median())
```

```python
for name, df in datasets.items():
    print(f"\n{name}")
    print(df.isnull().sum())
```

### Data Visualization

#### Bar chart

```python
plt.figure(figsize=(6,4))
bank_df["gender"].value_counts().plot(kind="bar")
plt.title("Customer Gender Distribution")
plt.xlabel("Gender")
plt.ylabel("Number of Customers")
plt.show()
```

```python
plt.figure(figsize=(8,5))
bank_df["employment_type"].value_counts().plot(kind="bar")
plt.title("Employment Type Distribution")
plt.xlabel("Employment Type")
plt.ylabel("Number of Customers")
plt.show()
```

```python
plt.figure(figsize=(6,4))
bank_df["payment_status"].value_counts().plot(kind="bar")
plt.title("Payment Status Distribution")
plt.xlabel("Payment Status")
plt.ylabel("Number of Payments")
plt.show()
```

#### histogram chart

```python
plt.figure(figsize=(8,4))
plt.hist(bank_df["annual_income"], bins=20)
plt.title("Annual Income Distribution")
plt.xlabel("Annual Income")
plt.ylabel("Frequency")
plt.show()
```

```python
plt.figure(figsize=(8,4))
plt.hist(bank_df["cibil_score"], bins=20)
plt.title("CIBIL Score Distribution")
plt.xlabel("CIBIL Score")
plt.ylabel("Frequency")
plt.show()
```

#### Pie chart

```python
plt.figure(figsize=(6,6))
bank_df["loan_status"].value_counts().plot(kind="pie", autopct="%1.1f%%")
plt.title("Loan Status")
plt.ylabel("")
plt.show()
```

```python
plt.figure(figsize=(6,6))
bank_df["payment_status"].value_counts().plot(kind="pie", autopct="%1.1f%%")
plt.title("Payment Status")
plt.ylabel("")
plt.show()
```
